```python
#Import Library
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib import colors
from matplotlib.ticker import PercentFormatter

#Baca dataset
dataset = pd.read_csv('https://raw.githubusercontent.com/ELDANIARA/KG-UAS-PRAK/main/stock.csv')
dataset.head(13)
```

# grafik histogram atribut Purchase Price, Current Price dan Gain/Loss

```python
plt.figure(figsize=(25, 10))

plt.subplot(1, 3, 1)
sns.histplot(dataset['Purchase Price ($)'], kde=True, bins=30, edgecolor='k', color="magenta")
plt.xlabel('Purchase Price', fontsize=15, color='magenta')
plt.ylabel('Frequency')
plt.title('Histogram of Purchase Price', fontsize=15)

plt.subplot(1, 3, 2)
sns.histplot(dataset['current Price ($)'], kde=True, bins=30, edgecolor='k', color="blue")
plt.xlabel('current Price', fontsize=15, color='blue')
plt.ylabel('Frequency')
plt.title('Histogram of current Price', fontsize=15)

plt.subplot(1, 3, 3)
sns.histplot(dataset['Gain/Loss ($)'], kde=True, bins=30, edgecolor='k', color="brown")
plt.xlabel('Gain/Loss', fontsize=15, color='brown')
plt.ylabel('Frequency')
plt.title('Histogram of Gain/Loss', fontsize=15)

plt.tight_layout()
plt.show()
```

# visualisasi data untuk menampilkan Atribut Stock Name dan Symbol yang memiliki Dividen / Share 5 tertinggi

```python
#Distribusi dataset pada setiap atribut Purchase Price, Current Price, dan Gain/Loss dengan grafik histogram
#Mengurutkan data berdasarkan dari Dividend / Shares
data = dataset.sort_values(by="Dividend/share", ascending=False)

#mendapatkan 5 Dividen/Shares tertinggi
top_5 = data.head(5)

#Plot the Data
plt.figure(figsize=(12, 5))
sns.barplot(x="Stock Name", y="Dividend/share", data=top_5)
plt.title("5 Saham Teratas dengan Dividen/share Tertinggi", loc='center',pad=10, fontsize=15, color='violet')
plt.show()
```

```python
#Distribusi dataset pada setiap atribut Purchase Price, Current Price, dan Gain/Loss dengan grafik histogram
#Mengurutkan data berdasarkan dari Dividend / Shares
data = dataset.sort_values(by="Dividend/share", ascending=False)

#mendapatkan 5 Dividen/Shares tertinggi
top_5 = data.head(5)

#Plot the Data
plt.figure(figsize=(10, 4))
sns.histplot(x="Stock Name", hue="Dividend/share", multiple="stack", data=top_5)
plt.title("5 Saham Teratas dengan Dividen/share Tertinggi", loc='center',pad=10, fontsize=15, color='violet')
plt.show()
```

# visualisasi data untuk menampilkan Atribut Stock Name dan Symbol yang memiliki Market Value 5 tertinggi

```python
#Mengurutkan data berdasarkan dari Market Value
data = dataset.sort_values(by="Market Value ($)", ascending=False)

#Mendapatkan 5 Market Value tertinggi
top_5 = data.head(5)

# Plot the data
plt.figure(figsize=(12, 5))
sns.barplot(x="Stock Name", y="Market Value ($)", data=top_5)
plt.title("5 Saham Teratas dengan Nilai Pasar Tertinggi", loc='center',pad=10, fontsize=15, color='violet')
plt.show()
```
